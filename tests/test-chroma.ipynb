```python
import chromadb

# Connect to the Chroma server running on localhost:8000
client = chromadb.HttpClient(host="localhost", port=8000)

# List all databases
databases = client.list_collections()

print("Chroma Databases/Collections on localhost:8000:")
print(f"Total collections: {len(databases)}")
print("\nCollections:")
for collection in databases:
    print(f"  - {collection.name}")
    print(f"    Metadata: {collection.metadata}")
    print(f"    Document count: {collection.count()}")
```
